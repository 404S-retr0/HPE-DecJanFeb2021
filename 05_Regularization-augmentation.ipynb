```
import Augmentor
pipes = Augmentor.Pipeline('mantisshrimp')
# Pizza pipeline
# 1 person-> Call, order, baking, packing, delivery
# pipeline-> 1 person(call),...., 1(delivery)
# 100 calls, 5 pizzas at a time, 2 deliveries
# SCALED people individually!
```

```
pipes.flip_left_right(0.5)
pipes.black_and_white(0.1)
pipes.random_distortion(1, 10,10, 1)
pipes.rotate(0.5, 25, 25)
pipes.skew(0.25, 0.25)
pipes.sample(5)
```
